```python
#creo un dataframe a partir del archivo csv 
df= pd.read_csv('./Datasets/precios_semanas_20200419_20200426.csv')
df
```

```python
# modifico el formato de la tabla producto_id eliminando la notacion cientifica 
df['producto_id']=df['producto_id'].astype(str).str.replace('e+' ,"")
```

```python
df
```

```python
#elimino el cero despues del punto en la tabla producto_id
df['producto_id'] = df['producto_id'].astype('string').str.rstrip('.0')
```

```python
# estandarizo la cantidad de datos en la columa producto_id para que todos cuenten con trece digitos 
df['producto_id']=df['producto_id'].apply(lambda x: '{0:0>13}'.format(x))
```

```python
#reordeno la pocision de las columnas del dataframe para que coincida con los creados posteriormente 
df = df[['precio','producto_id','sucursal_id']]
df
```
